4.  ë‘ ëª¨ë¸ì˜ ê²°ê³¼ë¥¼ ë¹„êµí•˜ê³ , 'ì•„ì´ìŠ¤í¬ë¦¼ íŒë§¤ëŸ‰'ì˜ ê³„ìˆ˜(coef)ì™€ p-valueê°€ ì–´ë–»ê²Œ ë³€í–ˆëŠ”ì§€ í•´ì„í•´ ë³´ì„¸ìš”.


```python
# 1. ê°€ìƒ ë°ì´í„° ìƒì„± (ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”)
np.random.seed(42)
n_samples = 100
temperature = np.random.uniform(15, 35, n_samples)
ice_cream_sales = 2 * temperature + np.random.normal(0, 5, n_samples)
beach_visitors = 10 * temperature + np.random.normal(0, 20, n_samples)
df1 = pd.DataFrame({'ê¸°ì˜¨': temperature, 'ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰': ice_cream_sales, 'í•´ìˆ˜ìš•ê°ìˆ˜': beach_visitors})
print(df1.corr())

# 2. ë‹¨ìˆœíšŒê·€ë¶„ì„: ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰ -> í•´ìˆ˜ìš•ê°ìˆ˜
print("\n--- ëª¨ë¸ 1: ë‹¨ìˆœíšŒê·€ë¶„ì„ (ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰ -> í•´ìˆ˜ìš•ê°ìˆ˜) ---")
y = df1['í•´ìˆ˜ìš•ê°ìˆ˜']
# ë…ë¦½ë³€ìˆ˜ë¡œ 'ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰'ë§Œ ì„ íƒí•˜ì„¸ìš”.
X_simple = df1[['ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰']]
# X_simpleì— ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ì„¸ìš”.
X_simple_const = sm.add_constant(X_simple)
# OLS ëª¨ë¸ì„ yì™€ X_simple_constë¡œ í•™ìŠµì‹œí‚¤ì„¸ìš”.
model_simple = sm.OLS(y, X_simple_const).fit()
print(model_simple.summary())

# 3. ë‹¤ì¤‘íšŒê·€ë¶„ì„: (ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰, ê¸°ì˜¨) -> í•´ìˆ˜ìš•ê°ìˆ˜
print("\n--- ëª¨ë¸ 2: ë‹¤ì¤‘íšŒê·€ë¶„ì„ (ì¤‘ì²©ìš”ì¸ 'ê¸°ì˜¨' í†µì œ) ---")
# ë…ë¦½ë³€ìˆ˜ë¡œ 'ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰'ê³¼ 'ê¸°ì˜¨'ì„ ëª¨ë‘ ì„ íƒí•˜ì„¸ìš”.
X_multi = df1[['ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰', 'ê¸°ì˜¨']]
# X_multiì— ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ì„¸ìš”.
X_multi_const = sm.add_constant(X_multi)
# OLS ëª¨ë¸ì„ yì™€ X_multi_constë¡œ í•™ìŠµì‹œí‚¤ì„¸ìš”.
model_multi = sm.OLS(y,  X_multi_const).fit()
print(model_multi.summary())

# 4. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# ëª¨ë¸ 1ì—ì„œ ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰ì˜ ê³„ìˆ˜ì™€ p-value:  ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰ì˜ ê³„ìˆ˜: 4.5550   p-value:0.000
# ëª¨ë¸ 2ì—ì„œ ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰ì˜ ê³„ìˆ˜ì™€ p-value:  ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰ì˜ ê³„ìˆ˜: -0.1087   p-value:0.826
# ê²°ê³¼ê°€ ë³€í•œ ì´ìœ : ë‹¨ìˆœíšŒê·€ë¶„ì„ì—ì„œëŠ” ê¸°ì˜¨ì´ë¼ëŠ” ì¤‘ì²©ìš”ì¸ì„ ê³ ë ¤í•˜ì§€ ì•Šì•˜ë‹¤. ë‹¤ì¤‘íšŒê·€ë¶„ì„ì—ì„œ ê¸°ì˜¨ì´ë¼ëŠ” ì¤‘ì²©ìš”ì¸ì„ ê³ ë ¤í•˜ì—¬ í†µì œí•˜ì
# ì•„ì´ìŠ¤í¬ë¦¼ íŒë§¤ëŸ‰ì˜ ê³„ìˆ˜ëŠ” 0ì— ê°€ê¹Œì›Œì§€ê³  pê°’ì´ ì¦ê°€í•˜ì—¬ í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•˜ì§€ ì•Šê²Œ ë˜ì—ˆë‹¤.
```

## ë¬¸ì œ 2 (ë‚œì´ë„: í•˜): í‘œë³¸ í¬ê¸°ì˜ í•¨ì •ê³¼ íš¨ê³¼ í¬ê¸°ì˜ ì¤‘ìš”ì„±

> **ğŸ¯ ëª©í‘œ:** í‘œë³¸ í¬ê¸°(n)ê°€ ì»¤ì§€ë©´ ì•„ì£¼ ë¯¸ë¯¸í•œ ì°¨ì´ë„ í†µê³„ì ìœ¼ë¡œ ìœ ì˜(p<0.05)í•˜ê²Œ ë˜ëŠ” í˜„ìƒì„ í™•ì¸í•˜ê³ , ì´ ë•Œë¬¸ì— **íš¨ê³¼ í¬ê¸°(Effect Size)**ë¥¼ í•¨ê»˜ í™•ì¸í•´ì•¼ í•˜ëŠ” ì´ìœ ë¥¼ ì´í•´í•©ë‹ˆë‹¤.


### ğŸ’¡ í•µì‹¬ ê°œë…:

ê°€ì„¤ê²€ì •ì—ì„œ p-valueëŠ” 'ê·€ë¬´ê°€ì„¤ì´ ë§ë‹¤ê³  ê°€ì •í•  ë•Œ, í˜„ì¬ ë°ì´í„°ì™€ ê°™ê±°ë‚˜ ë” ê·¹ë‹¨ì ì¸ ê²°ê³¼ê°€ ë‚˜ì˜¬ í™•ë¥ 'ì…ë‹ˆë‹¤. í‘œë³¸ì´ ì»¤ì§ˆìˆ˜ë¡ ì•„ì£¼ ì‘ì€ ì°¨ì´ë„ ìš°ì—°ìœ¼ë¡œ ë³´ê¸° ì–´ë ¤ì›Œì§€ë¯€ë¡œ p-valueëŠ” ì‘ì•„ì§‘ë‹ˆë‹¤. **íš¨ê³¼ í¬ê¸°**ëŠ” ì´ ì°¨ì´ê°€ 'ì‹¤ì§ˆì ìœ¼ë¡œ ì–¼ë§ˆë‚˜ ì˜ë¯¸ ìˆëŠ” í¬ê¸°ì¸ê°€'ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì§€í‘œë¡œ, p-valueì˜ í•œê³„ë¥¼ ë³´ì™„í•´ ì¤ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1.  í‰ê· ì´ ì•„ì£¼ ì•½ê°„ ë‹¤ë¥¸(ì˜ˆ: 170cm vs 170.1cm) ë‘ ì§‘ë‹¨ì„ ê°€ì •í•©ë‹ˆë‹¤.
2.  **ì‘ì€ í‘œë³¸(n=30)**ì„ ê° ì§‘ë‹¨ì—ì„œ ì¶”ì¶œí•˜ì—¬ t-ê²€ì •ì„ ìˆ˜í–‰í•˜ê³  p-valueë¥¼ í™•ì¸í•©ë‹ˆë‹¤.
3.  **í° í‘œë³¸(n=10,000)**ì„ ê° ì§‘ë‹¨ì—ì„œ ì¶”ì¶œí•˜ì—¬ t-ê²€ì •ì„ ìˆ˜í–‰í•˜ê³  p-valueë¥¼ í™•ì¸í•©ë‹ˆë‹¤.
4.  ë‘ ê²½ìš° ëª¨ë‘ì— ëŒ€í•´ **íš¨ê³¼ í¬ê¸°(Cohen's d)**ë¥¼ ì§ì ‘ ê³„ì‚°í•˜ê³ , p-valueì™€ íš¨ê³¼ í¬ê¸°ê°€ ì–´ë–»ê²Œ ë‹¤ë¥¸ ë©”ì‹œì§€ë¥¼ ì£¼ëŠ”ì§€ ë¹„êµ ë¶„ì„í•˜ì„¸ìš”.


```python
# íš¨ê³¼ í¬ê¸°(Cohen's d) ê³„ì‚° í•¨ìˆ˜ (ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”)
def cohen_d(x, y):
    nx, ny = len(x), len(y)
    dof = nx + ny - 2
    return (np.mean(x) - np.mean(y)) / np.sqrt(((nx-1)*np.std(x, ddof=1)**2 + (ny-1)*np.std(y, ddof=1)**2) / dof)

np.random.seed(123)
mu1, mu2, sigma = 170, 170.1, 5

# 2. ì‘ì€ í‘œë³¸(n=30)ìœ¼ë¡œ ê²€ì •
sample1_small = np.random.normal(mu1, sigma, 30)
sample2_small = np.random.normal(mu2, sigma, 30)

# scipy.stats.ttest_indë¥¼ ì‚¬ìš©í•˜ì—¬ ë‘ ì‘ì€ í‘œë³¸ ê°„ì˜ t-ê²€ì •ì„ ìˆ˜í–‰í•˜ì„¸ìš”.
t_stat_small, p_value_small = stats.ttest_ind(sample1_small, sample2_small)
effect_size_small = cohen_d(sample1_small, sample2_small)

print(f"--- ì‘ì€ í‘œë³¸ (n=30) ê²°ê³¼ ---")
print(f"P-value: {p_value_small:.4f}")
print(f"Effect Size (Cohen's d): {effect_size_small:.4f}\n")

# 3. í° í‘œë³¸(n=10000)ìœ¼ë¡œ ê²€ì •
sample1_large = np.random.normal(mu1, sigma, 10000)
sample2_large = np.random.normal(mu2, sigma, 10000)

# scipy.stats.ttest_indë¥¼ ì‚¬ìš©í•˜ì—¬ ë‘ í° í‘œë³¸ ê°„ì˜ t-ê²€ì •ì„ ìˆ˜í–‰í•˜ì„¸ìš”.
t_stat_large, p_value_large = stats.ttest_ind(sample1_large, sample2_large)
effect_size_large = cohen_d(sample1_large, sample2_large)

print(f"--- í° í‘œë³¸ (n=10000) ê²°ê³¼ ---")
print(f"P-value: {p_value_large:.4f}")
print(f"Effect Size (Cohen's d): {effect_size_large:.4f}\n")

# 4. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# í‘œë³¸ í¬ê¸°ê°€ ì»¤ì§€ì p-valueëŠ” ì–´ë–»ê²Œ ë³€í–ˆë‚˜ìš”?: í‘œë³¸í¬ê¸°ê°€ ì»¤ì§€ì P-valueì˜ ìˆ˜ì¹˜ê°€ 0.7100ì—ì„œ P-value: 0.1673ìœ¼ë¡œ ê°ì†Œí•˜ì˜€ë‹¤. ì´ëŠ” í‘œë³¸ ìˆ˜ê°€ ì¦ê°€í•˜ë©´ì„œ ê²€ì •ë ¥ì´ ì»¤ì ¸ ì‘ì€ í‰ê·  ì°¨ì´ë„ ì˜ ê°ì§€í•  ìˆ˜ ìˆê²Œ ë˜ê¸° ë•Œë¬¸ì´ë‹¤.
# ë‘ ê²½ìš°ì˜ íš¨ê³¼ í¬ê¸°ëŠ” ë¹„ìŠ·í•œê°€ìš”, ë‹¤ë¥¸ê°€ìš”?: ì´ ê²½ìš° íš¨ê³¼ì˜ í¬ê¸° ë˜í•œ Effect Size (Cohen's d): -0.0965ì—ì„œ Effect Size (Cohen's d): -0.0195ìœ¼ë¡œ ë‚®ì•„ì¡Œì§€ë§Œ, ë¯¸ë¯¸í•œ íš¨ê³¼ì˜ ì°¨ì´ë¥¼ ë³´ì˜€ë‹¤.

# ì§ˆë¬¸ : ê·¸ë ‡ë‹¤ë©´, í° í‘œë³¸ì¸ ê²½ìš° íš¨ê³¼í¬ê¸°ë¥¼ ê²€ì •í•˜ëŠ”ê²Œ í•„ìˆ˜ì ì¸ ê±´ê°€??? íš¨ê³¼í¬ê¸°ë¥¼ ê²€ì •í•´ì•¼ í•˜ëŠ” ë°ì´í„°ì˜ ì¡°ê±´ì´ ìˆëŠ” ê²ƒì¸ê°€???
```

### ğŸ¤” ìƒê°í•´ ë³¼ ë¬¸ì œ:

A/B í…ŒìŠ¤íŠ¸ì—ì„œ ì›¹ì‚¬ì´íŠ¸ ë²„íŠ¼ ìƒ‰ê¹”ì„ ë°”ê¿¨ë”ë‹ˆ, í´ë¦­ë¥ ì´ 0.1%p ì¦ê°€í–ˆê³ , p-valueëŠ” 0.001ì´ì—ˆìŠµë‹ˆë‹¤. ì´ ê²°ê³¼ë¥¼ ë³´ê³  ë²„íŠ¼ ìƒ‰ê¹”ì„ ë°”ê¾¸ëŠ” ê²ƒì´ í•­ìƒ ì˜³ë‹¤ê³  í•  ìˆ˜ ìˆì„ê¹Œìš”? 4.  **ì§ì§€ì–´ì§„ í›„ ë¹„êµ**: 'ì§ì§€ì–´ì§„' ìˆ˜ê°•ìƒ ì§‘ë‹¨ê³¼ ë¹„ìˆ˜ê°•ìƒ ì§‘ë‹¨ì˜ í‰ê·  ì„±ì ì„ ë¹„êµí•˜ê³ , ê·¸ ì°¨ì´(ATT: Average Treatment Effect on the Treated)ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
5.  ë‹¨ìˆœ ë¹„êµ ê²°ê³¼ì™€ PSM í›„ì˜ ê²°ê³¼ë¥¼ ë¹„êµí•˜ê³ , ì™œ ê²°ê³¼ê°€ ë‹¬ë¼ì¡ŒëŠ”ì§€ í•´ì„í•©ë‹ˆë‹¤.


```python
# 1. ì„ íƒ í¸í–¥ì´ ìˆëŠ” ë°ì´í„° ìƒì„±

# ë‚œìˆ˜ ê²°ê³¼ë¥¼ ê³ ì • (ê°™ì€ ê²°ê³¼ê°€ ë‚˜ì˜¤ë„ë¡ ì„¤ì •)
np.random.seed(42)

# ì „ì²´ í•™ìƒ ìˆ˜
n = 1000

# ì‚¬ì „ ì„±ì : í‰ê·  60, í‘œì¤€í¸ì°¨ 10ì¸ ì •ê·œë¶„í¬
pre_score = np.random.normal(60, 10, n)

# í•™ìŠµ ë™ê¸°: 0~1 ì‚¬ì´ì˜ ê· ë“±ë¶„í¬
motivation = np.random.uniform(0, 1, n)

# ì²˜ì¹˜ë¥¼ ë°›ì„ í™•ë¥ (ê²½í–¥ ì ìˆ˜)ì„ ê³„ì‚°
# ì‚¬ì „ ì„±ì ì´ ë†’ê³ , ë™ê¸°ê°€ ë†’ì„ìˆ˜ë¡ ì²˜ì¹˜ë¥¼ ë°›ì„ í™•ë¥ ì´ ì»¤ì§€ë„ë¡ ì„¤ê³„
propensity = 1 / (1 + np.exp(-( -5 + (pre_score-60)/5 + motivation*2)))

# ìœ„ì—ì„œ ê³„ì‚°í•œ í™•ë¥ ì„ ì´ìš©í•´ ì‹¤ì œ ì²˜ì¹˜ ì—¬ë¶€ë¥¼ ê²°ì • (0 ë˜ëŠ” 1)
treatment = np.random.binomial(1, propensity, n)

# ì‹¤ì œ ì²˜ì¹˜ íš¨ê³¼ (ì§„ì§œ ì¸ê³¼ íš¨ê³¼)
true_effect = 5

# ìµœì¢… ì„±ì  = ì‚¬ì „ ì„±ì  + ì²˜ì¹˜ íš¨ê³¼ + ì˜¤ì°¨
final_score = pre_score + treatment * true_effect + np.random.normal(0, 5, n)

# ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ì •ë¦¬
df4 = pd.DataFrame({
    'pre_score': pre_score,
    'motivation': motivation,
    'treatment': treatment,
    'final_score': final_score
})


# 2. ë‹¨ìˆœ ë¹„êµ (Naive Effect)

# ì²˜ì¹˜ ì§‘ë‹¨ê³¼ ë¹„ì²˜ì¹˜ ì§‘ë‹¨ì˜ í‰ê·  ì„±ì  ì°¨ì´ ê³„ì‚°
# (ì•„ë¬´ ë³´ì • ì—†ì´ ë‹¨ìˆœ ë¹„êµ)
naive_effect = (
    df4[df4['treatment'] == 1]['final_score'].mean()
    - df4[df4['treatment'] == 0]['final_score'].mean()
)

print(f"ë‹¨ìˆœ í‰ê·  ë¹„êµ (ì„±ì  ì°¨ì´): {naive_effect:.2f}ì \n")


# 3. ê²½í–¥ ì ìˆ˜(propensity score) ê³„ì‚°(ë¹„ìŠ·í•œ ì‚¬ëŒë“¤ì„ ë¬¶ê¸° ìœ„í•´ ìˆ˜í–‰)

# ì²˜ì¹˜ ì—¬ë¶€ë¥¼ ì„¤ëª…í•˜ëŠ” ë³€ìˆ˜ë“¤ (ê³µë³€ëŸ‰)
X = df4[['pre_score', 'motivation']]

# ì¢…ì†ë³€ìˆ˜: ì²˜ì¹˜ ì—¬ë¶€
y = df4['treatment']

# ë¡œì§€ìŠ¤í‹± íšŒê·€ë¥¼ ì´ìš©í•´ ì²˜ì¹˜ í™•ë¥ ì„ ì¶”ì •
logit_model = sm.Logit(y, sm.add_constant(X)).fit(disp=0)

# ê° ê°œì¸ì˜ ê²½í–¥ ì ìˆ˜(ì²˜ì¹˜ ë°›ì„ í™•ë¥ )ë¥¼ ë°ì´í„°í”„ë ˆì„ì— ì €ì¥
df4['propensity_score'] = logit_model.predict(sm.add_constant(X))

# 3. ê²½í–¥ ì ìˆ˜ ë§¤ì¹­ (Matching)

# ì²˜ì¹˜ ì§‘ë‹¨ê³¼ í†µì œ ì§‘ë‹¨ì„ ë¶„ë¦¬
treated = df4[df4['treatment'] == 1]
control = df4[df4['treatment'] == 0]

# ìµœê·¼ì ‘ ì´ì›ƒ ë§¤ì¹­ ì„¤ì •
# n_neighbors=1 : ê°€ì¥ ê°€ê¹Œìš´ 1ëª…ê³¼ ë§¤ì¹­
# ê²½í–¥ ì ìˆ˜ê°€ ê°€ì¥ ë¹„ìŠ·í•œ í†µì œ ì§‘ë‹¨ 1ëª…ì„ ì°¾ì•„ 1:1ë¡œ ì§ì§€ì–´ ì£¼ëŠ” ê³¼ì •
# radius=0.5ëŠ” ë„ˆë¬´ ë©€ë©´ ë§¤ì¹­í•˜ì§€ ì•Šê² ë‹¤
nn = NearestNeighbors(n_neighbors=1, radius=0.5)

# í†µì œ ì§‘ë‹¨ì˜ ê²½í–¥ ì ìˆ˜ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ëª¨ë¸ í•™ìŠµ
nn.fit(control[['propensity_score']])

# ì²˜ì¹˜ ì§‘ë‹¨ì˜ ê° ì‚¬ëŒì— ëŒ€í•´
# ê²½í–¥ ì ìˆ˜ê°€ ê°€ì¥ ê°€ê¹Œìš´ í†µì œ ì§‘ë‹¨ ì‚¬ëŒ ì°¾ê¸°
distances, indices = nn.kneighbors(treated[['propensity_score']])

# ë§¤ì¹­ëœ í†µì œ ì§‘ë‹¨ ë°ì´í„° ì„ íƒ
matched_control = control.iloc[indices.flatten()]


# 4. ì§ì§€ì–´ì§„ í›„ ë¹„êµ (ATT ê³„ì‚°:"ì²˜ì¹˜ë¥¼ ì‹¤ì œë¡œ ë°›ì€ ì‚¬ëŒë“¤ì—ê²Œì„œ ê·¸ ì²˜ì¹˜ê°€ í‰ê· ì ìœ¼ë¡œ ì–¼ë§ˆë‚˜ íš¨ê³¼ê°€ ìˆì—ˆëŠ”ì§€ë¥¼ í™•ì¸í•˜ëŠ” ìœ„í•´ ATTë¥¼ ì‚¬ìš© ")
# PSMì—ì„œëŠ” ATTë¥¼ ê°€ì¥ ë§ì´ ê³„ì‚°í•¨.

# ì²˜ì¹˜ ì§‘ë‹¨ê³¼ ë§¤ì¹­ëœ í†µì œ ì§‘ë‹¨ì˜ í‰ê·  ì„±ì  ì°¨ì´
# â†’ ATT (ì²˜ì¹˜ë¥¼ ë°›ì€ ì‚¬ëŒì—ê²Œì„œì˜ í‰ê·  ì²˜ì¹˜ íš¨ê³¼)
att_effect = (
    treated['final_score'].mean()
    - matched_control['final_score'].mean()
)

print(f"PSM ì´í›„ í‰ê·  ë¹„êµ (ì¶”ì •ëœ ì¸ê³¼ íš¨ê³¼): {att_effect:.2f}ì \n")


# 5. ê²°ê³¼ í•´ì„ (ê°œë…)

# ì§ˆë¬¸1: ë‹¨ìˆœ ë¹„êµ ê²°ê³¼ê°€ ì‹¤ì œ íš¨ê³¼(5ì )ë³´ë‹¤ í¬ê²Œ ë‚˜ì˜¨ ì´ìœ ëŠ”?:
# ì²˜ì¹˜ ì§‘ë‹¨ì€ ì›ë˜ ì‚¬ì „ ì„±ì ê³¼ ë™ê¸°ê°€ ë†’ì€ ì‚¬ëŒë“¤ì´ ë” ë§ì´ í¬í•¨ë˜ì–´ ìˆì–´, ë‹¨ìˆœ ë¹„êµì—ì„œëŠ” ì²˜ì¹˜ íš¨ê³¼ì™€ ê¸°ì¡´ ëŠ¥ë ¥ ì°¨ì´ê°€ í•¨ê»˜ ì„ì—¬ ì¶”ì •ë˜ì—ˆê¸° ë•Œë¬¸ì´ë‹¤.
# ì‰½ê²Œ ë§í•´ ì²˜ì¹˜ ì§‘ë‹¨ì˜ í•™ìƒë“¤ì€ ì›ë˜ ì„±ì ë„ ë†’ê³  ê³µë¶€ì˜ì§€ë„ ë†’ì•˜ê¸°ì—, ì„±ì ì´ ë†’ì€ ì´ìœ ê°€ ë­ ë•Œë¬¸ì´ì—ˆëŠ”ì§€ êµ¬ë¶„í•˜ê¸°ê°€ ë§¤ìš° ì–´ë µë‹¤. ê·¸ë˜ì„œ ë„ì¶œëœ íš¨ê³¼ì„±ì„ ê³¼ëŒ€í‰ê°€í•˜ê²Œ ë˜ëŠ”ë°, ì´ê²Œ ì„ íƒ í¸í–¥ì´ë‹¤.

# ì§ˆë¬¸2: PSMì„ í†µí•´ ì´ ë¬¸ì œë¥¼ ì–´ë–»ê²Œ í•´ê²°í•  ìˆ˜ ìˆì—ˆë‚˜ìš”?:
# PSMì€ ì‚¬ì „ ì„±ì ê³¼ ë™ê¸°ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì²˜ì¹˜ë¥¼ ë°›ì„ í™•ë¥ ì´ ë¹„ìŠ·í•œ ì‚¬ëŒë¼ë¦¬ ë¹„êµí•˜ì—¬(ì‚¬ì „ ì„±ì ë„ ë¹„ìŠ·í•˜ê³  ê³µë¶€ì˜ì§€ë„ ë¹„ìŠ·í•œ í•™ìƒë¼ë¦¬ êµ¬ë¶„í•˜ì—¬ ë¬¶ì–´ì„œ ë¹„êµí•˜ëŠ” ê²ƒì´ë‹¤.), ì²˜ì¹˜ ì—¬ë¶€ë¥¼ ì œì™¸í•œ ë‹¤ë¥¸ ìš”ì¸ì˜ ì˜í–¥ì„ ì¤„ì´ê³ , ì‹¤ì œ ì¸ê³¼ íš¨ê³¼ì— ë” ê°€ê¹Œìš´ ê°’ì„ ì¶”ì •í•  ìˆ˜ ìˆê²Œ í•´ì¤€ë‹¤.

# ê²°ê³¼í•´ì„: ë‹¨ìˆœë¹„êµëŠ” ì‚¬ê³¼ì™€ ì˜¤ë Œì§€ë¥¼ ë¹„êµí•œë‹¤ë©´, psmì€ ì‚¬ê³¼ì™€ ì‚¬ê³¼ë¥¼ ë¹„êµí•œë‹¬ê¹Œ.
# ë‹¨ìˆœ í‰ê·  ë¹„êµ ê²°ê³¼ ì²˜ì¹˜ ì§‘ë‹¨ì´ í†µì œ ì§‘ë‹¨ë³´ë‹¤ ì•½ 19.96 ë†’ê²Œ ë‚˜ì™”ë‹¤. ê·¸ëŸ¬ë‚˜ ì´ ê°’ì€ ì²˜ì¹˜ ì§‘ë‹¨ì´ ì›ë˜ ì‚¬ì „ ì„±ì ê³¼ ë™ê¸°ê°€ ë” ë†’ì€ ì‚¬ëŒë“¤ë¡œ êµ¬ì„±ëœ ì˜í–¥ì´ í•¨ê»˜ í¬í•¨ëœ ê²°ê³¼ì´ë‹¤.
# psm ì´í›„ í‰ê· ì„ ë¹„êµí•˜ë‹ˆ, ì¶”ì²­ëœ ì¸ê³¼ íš¨ê³¼ê°€ ì‹¤ì œ íš¨ê³¼(5ì )ì— ê°€ê¹Œìš´ 4.31ë¡œ ë‚˜íƒ€ë‚¬ë‹¤.
```

ì§ì§“ê¸°(Matching)ê°€ ì˜ ë˜ì—ˆëŠ”ì§€ í™•ì¸í•˜ëŠ” ë°©ë²• ì¤‘ í•˜ë‚˜ëŠ”, ì§ì§€ì–´ì§„ ë‘ ì§‘ë‹¨ì˜ ì¤‘ì²©ìš”ì¸(ì—¬ê¸°ì„œëŠ” `pre_score`, `motivation`) í‰ê· ì´ ì„œë¡œ ë¹„ìŠ·í•´ì¡ŒëŠ”ì§€ í™•ì¸í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤. ì´ë¥¼ 'ê³µë³€ëŸ‰ ê· í˜•(Covariate Balance)' í™•ì¸ì´ë¼ê³  í•©ë‹ˆë‹¤. ì§ì§“ê¸° ì „ê³¼ í›„ì˜ `pre_score` í‰ê·  ì°¨ì´ë¥¼ ê°ê° ê³„ì‚°í•´ë³´ê³ , PSMì´ ê³µë³€ëŸ‰ ê· í˜•ì„ ë§ì¶”ëŠ” ë° ì„±ê³µí–ˆëŠ”ì§€ í™•ì¸í•´ë³´ì„¸ìš”.



```python
# ê³µë³€ëŸ‰ ê· í˜•(Covariate Balance) í™•ì¸
# ê³µë³€ëŸ‰ ê· í˜•ì´ë€> ê³µë³€ëŸ‰ì€ ì²˜ì¹˜ì™€ ê²°ê³¼ ëª¨ë‘ì— ì˜í–¥ì„ ì¤„ ìˆ˜ ìˆëŠ” ê¸°ì¡´ íŠ¹ì„±ì˜ ë³€ìˆ˜ë¥¼ ì˜ë¯¸í•œë‹¤.

# ì§ì§“ê¸° ì „  pre_score í‰ê·  ì°¨ì´
pre_score_diff_before = (
    df4[df4['treatment'] == 1]['pre_score'].mean()
    - df4[df4['treatment'] == 0]['pre_score'].mean()
)

print(f"ì§ì§“ê¸° ì „ pre_score í‰ê·  ì°¨ì´: {pre_score_diff_before:.2f}")

# ì§ì§“ê¸° í›„  pre_score í‰ê·  ì°¨ì´
pre_score_diff_after = (
    treated['pre_score'].mean()
    - matched_control['pre_score'].mean()
)

print(f"ì§ì§“ê¸° í›„ pre_score í‰ê·  ì°¨ì´: {pre_score_diff_after:.2f}")
```

### ğŸ’¡ í•µì‹¬ ê°œë…:

**ì´ì¤‘ì°¨ë¶„ë²•(DiD)**ì€ ì •ì±…ì´ ì‹œí–‰ëœ **ì‹¤í—˜êµ°(Treatment Group)**ê³¼ ì‹œí–‰ë˜ì§€ ì•Šì€ **í†µì œêµ°(Control Group)**ì˜ **ì •ì±… ì‹œí–‰ ì „í›„(Pre-Post)** ë°ì´í„°ë¥¼ ëª¨ë‘ í™œìš©í•˜ëŠ” ê°•ë ¥í•œ ì¸ê³¼ì¶”ë¡  ë°©ë²•ì…ë‹ˆë‹¤. í•µì‹¬ ì•„ì´ë””ì–´ëŠ” 'ì •ì±…ì´ ì—†ì—ˆì„ ê²½ìš°ì˜ ìì—°ì ì¸ ì‹œê°„ ë³€í™”'ë¥¼ í†µì œêµ°ì„ í†µí•´ ì¶”ì •í•˜ê³ , ì´ë¥¼ ì‹¤í—˜êµ°ì˜ ë³€í™”ì—ì„œ ë¹¼ ì¤Œìœ¼ë¡œì¨ ì •ì±…ì˜ ìˆœìˆ˜í•œ íš¨ê³¼ë§Œ ë°œë¼ë‚´ëŠ” ê²ƒì…ë‹ˆë‹¤. íšŒê·€ë¶„ì„ì—ì„œëŠ” `(ì‹¤í—˜êµ° ì—¬ë¶€) * (ì •ì±… í›„ ì—¬ë¶€)`ì˜ **ìƒí˜¸ì‘ìš© í•­** ê³„ìˆ˜ê°€ ë°”ë¡œ DiD ì¶”ì •ì¹˜ê°€ ë©ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1.  'ì„œìš¸ì‹œ'ì—ë§Œ 'ë¬´ë£Œ ê³µê³µ ì™€ì´íŒŒì´' ì •ì±…ì„ ë„ì…í–ˆì„ ë•Œ, ì‹œë¯¼ë“¤ì˜ 'ì›” í‰ê·  ë°ì´í„° ì‚¬ìš©ëŸ‰'ì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì„ ë¶„ì„í•˜ëŠ” ê°€ìƒ ë°ì´í„°ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. (ì‹¤í—˜êµ°: ì„œìš¸, í†µì œêµ°: ë¶€ì‚°)
2.  **ì‹œê°í™”**: ì •ì±… ì‹œí–‰ ì „í›„, ë‘ ë„ì‹œì˜ í‰ê·  ë°ì´í„° ì‚¬ìš©ëŸ‰ ë³€í™”ë¥¼ êº¾ì€ì„  ê·¸ë˜í”„ë¡œ ê·¸ë ¤ë´…ë‹ˆë‹¤.
3.  **DiD íšŒê·€ë¶„ì„**: `ë°ì´í„°ì‚¬ìš©ëŸ‰ ~ is_ì„œìš¸ + is_ì •ì±…í›„ + is_ì„œìš¸*is_ì •ì±…í›„` í˜•íƒœì˜ íšŒê·€ ëª¨ë¸ì„ ë§Œë“­ë‹ˆë‹¤.
    - `is_ì„œìš¸`: ì„œìš¸ì´ë©´ 1, ë¶€ì‚°ì´ë©´ 0
    - `is_ì •ì±…í›„`: ì •ì±… ì‹œí–‰ í›„(2024ë…„)ì´ë©´ 1, ì‹œí–‰ ì „(2023ë…„)ì´ë©´ 0
    - `is_ì„œìš¸*is_ì •ì±…í›„`: ìƒí˜¸ì‘ìš© í•­
4.  íšŒê·€ë¶„ì„ ê²°ê³¼í‘œì—ì„œ ìƒí˜¸ì‘ìš© í•­ì˜ ê³„ìˆ˜(coef)ì™€ p-valueë¥¼ í™•ì¸í•˜ê³ , ì •ì±…ì˜ íš¨ê³¼ë¥¼ í•´ì„í•´ ë³´ì„¸ìš”.


```python
# 1. DiD ë¶„ì„ì„ ìœ„í•œ ê°€ìƒ íŒ¨ë„ ë°ì´í„° ìƒì„± (ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”)
np.random.seed(2025)
n_per_group = 500
df_seoul_pre = pd.DataFrame({'city': 'ì„œìš¸', 'year': 2023, 'data_usage': np.random.normal(15, 3, n_per_group)})
df_seoul_post = pd.DataFrame({'city': 'ì„œìš¸', 'year': 2024, 'data_usage': df_seoul_pre['data_usage'] + 2 + 5 + np.random.normal(0, 1, n_per_group)})
df_busan_pre = pd.DataFrame({'city': 'ë¶€ì‚°', 'year': 2023, 'data_usage': np.random.normal(10, 3, n_per_group)})
df_busan_post = pd.DataFrame({'city': 'ë¶€ì‚°', 'year': 2024, 'data_usage': df_busan_pre['data_usage'] + 2 + np.random.normal(0, 1, n_per_group)})
df5 = pd.concat([df_seoul_pre, df_seoul_post, df_busan_pre, df_busan_post])

# 2. ì‹œê°í™” (ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”)
sns.lineplot(data=df5, x='year', y='data_usage', hue='city', marker='o', errorbar='ci')
plt.title('ì •ì±… ì‹œí–‰ ì „í›„ ë„ì‹œë³„ í‰ê·  ë°ì´í„° ì‚¬ìš©ëŸ‰ ë³€í™”')
plt.xticks([2023, 2024]); plt.grid(True, axis='y', linestyle='--')
plt.show()

# 3. DiD íšŒê·€ë¶„ì„ì„ ìœ„í•œ ë³€ìˆ˜ ìƒì„±
# 'is_seoul' ë³€ìˆ˜: cityê°€ 'ì„œìš¸'ì´ë©´ 1, ì•„ë‹ˆë©´ 0
df5['is_seoul'] = (df5['city'] == 'ì„œìš¸').astype(int)
# 'is_post' ë³€ìˆ˜: yearê°€ 2024ì´ë©´ 1, ì•„ë‹ˆë©´ 0
df5['is_post'] = (df5['year'] == 2024).astype(int)
# ìƒí˜¸ì‘ìš© í•­ 'did_term': 'is_seoul'ê³¼ 'is_post'ì˜ ê³±
df5['did_term'] = df5['is_seoul'] * df5['is_post']

y = df5['data_usage']
# ë…ë¦½ë³€ìˆ˜ Xì— DiD ë¶„ì„ì— í•„ìš”í•œ ì„¸ ë³€ìˆ˜('is_seoul', 'is_post', 'did_term')ë¥¼ í¬í•¨ì‹œí‚¤ì„¸ìš”.
X = df5[['is_seoul', 'is_post', 'did_term']]
# ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ê³  OLS ëª¨ë¸ì„ í•™ìŠµì‹œí‚¤ì„¸ìš”.
model_did = sm.OLS(y, sm.add_constant(X)).fit()
print(model_did.summary())

# 4. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# is_seoul ê³„ìˆ˜ì˜ ì˜ë¯¸:is_seoul 4.6588
# is_post ê³„ìˆ˜ì˜ ì˜ë¯¸:is_post  2.0141
# did_term (ìƒí˜¸ì‘ìš© í•­) ê³„ìˆ˜ì˜ ì˜ë¯¸: did_term 4.9832
# ë”°ë¼ì„œ, ë¬´ë£Œ ì™€ì´íŒŒì´ ì •ì±…ì˜ ìˆœìˆ˜í•œ íš¨ê³¼ëŠ” ì•½ ëª‡ GB ì¦ê°€ë¼ê³  í•  ìˆ˜ ìˆë‚˜ìš”?:
# DiD ë¶„ì„ ê²°ê³¼, ë¬´ë£Œ ì™€ì´íŒŒì´ ì •ì±… ì‹œí–‰ ì´í›„ ì„œìš¸ì˜ ë°ì´í„° ì‚¬ìš©ëŸ‰ì€ ë¶€ì‚°ì˜ ë³€í™” ì¶”ì„¸ë¥¼ ê³ ë ¤í•˜ë”ë¼ë„ í‰ê· ì ìœ¼ë¡œ ì•½ 5GB ì¶”ê°€ ì¦ê°€í•œ ê²ƒìœ¼ë¡œ ë‚˜íƒ€ë‚¬ë‹¤.
```
