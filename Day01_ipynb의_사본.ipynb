```python
# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
np.random.seed(2025)
satisfaction = np.random.normal(7.5, 1.5, 1000)
satisfaction = np.clip(satisfaction, 1, 10)
df1 = pd.DataFrame({'score': satisfaction})
```

```python
# Ïó¨Í∏∞Ïóê ÏΩîÎìú ÏûëÏÑ±
mean_val = df1['score'].mean()
median_val = df1['score'].median()
mode_val = df1['score'].mode()
std_val = df1['score'].std()

mean_val, median_val, mode_val, std_val

plt.figure(figsize=(10, 6))
sns.histplot(df1['score'], bins=20, kde=True, color='skyblue')

# ÌèâÍ∑†, Ï§ëÏïôÍ∞í ÏÑ† Ï∂îÍ∞Ä
plt.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='-', linewidth=2, label=f'Median: {median_val:.2f}')
plt.axvline(mode_val.iloc[0], color='purple', linestyle='-.', linewidth=2, label=f'Mode: {mode_val.iloc[0]:.2f}')

plt.title('Customer Satisfaction Distribution')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(2025)
class1 = np.random.normal(75, 8, 30)
class2 = np.concatenate([np.random.normal(75, 15, 28),[0, 130]])
df3 = pd.DataFrame({'class': ['1class'] * 30 + ['2class'] * 30,'score': np.concatenate([class1, class2])})

stats = df3.groupby('class')['score'].agg(['mean', 'std', 'min', 'max'])
print("Statistics by class:\n", stats)

plt.figure(figsize=(6, 5))
df3.boxplot(by='class', column='score')
plt.title("Score Distribution by Class (Boxplot)")
plt.suptitle("")  # remove default title
plt.xlabel("Class")
plt.ylabel("Score")
plt.show()
```

# Î¨∏Ï†ú 4. IQR Ïù¥ÏÉÅÍ∞í Ï†úÍ±∞
**üìò Î¨∏Ï†ú**
- ÏïûÏùò Î¨∏Ï†úÏóêÏÑú 2Î∞òÏóêÎäî **Í∑πÎã®Ï†ÅÏù∏ Ïù¥ÏÉÅÍ∞í(Ïòà: Îß§Ïö∞ ÎÇÆÏùÄ Ï†êÏàò)**Ïù¥ Ìè¨Ìï®ÎêòÏñ¥ ÏûàÏñ¥
ÌÜµÍ≥Ñ ÏßÄÌëú, ÌäπÌûà ÌèâÍ∑†Ïóê ÌÅ∞ ÏòÅÌñ•ÏùÑ Ï£ºÎäî Í≤ÉÏùÑ ÌôïÏù∏ÌñàÏäµÎãàÎã§.
- Ïù¥Î≤àÏóêÎäî **IQR(ÏÇ¨Î∂ÑÏúÑ Î≤îÏúÑ)**ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÍ∞íÏùÑ Ï†úÍ±∞Ìïú Îí§, Ïù¥ÏÉÅÍ∞í Ï†úÍ±∞ Ï†ÑÌõÑÏùò ÌèâÍ∑†ÏùÑ ÎπÑÍµêÌï¥ Î¥ÖÏãúÎã§

<br>

**üìå ÏïÑÎûòÎ•º ÏàòÌñâÌï¥ Î≥¥ÏÑ∏Ïöî:**
- 2Î∞òÏùò Ï†êÏàò Îç∞Ïù¥ÌÑ∞Î•º ÎåÄÏÉÅÏúºÎ°ú IQR Í∏∞Ï§ÄÏùÑ ÏÇ¨Ïö©Ìï¥ Ïù¥ÏÉÅÍ∞íÏùÑ ÌåêÎ≥ÑÌïòÍ≥† Ï†úÍ±∞Ìï¥ Î¥ÖÏãúÎã§.
- Ïù¥ÏÉÅÍ∞í Ï†úÍ±∞ Ï†ÑÍ≥º Ï†úÍ±∞ ÌõÑÏùò ÌèâÍ∑†ÏùÑ Í∞ÅÍ∞Å Í≥ÑÏÇ∞Ìï¥ Î¥ÖÏãúÎã§.
- Ïù¥ÏÉÅÍ∞í Ï†úÍ±∞Í∞Ä ÌèâÍ∑†Ïóê Ïñ¥Îñ§ ÏòÅÌñ•ÏùÑ Ï£ºÎäîÏßÄ ÏÑ§Î™ÖÌï¥ Î¥ÖÏãúÎã§.
- ÌïÑÏöîÌïòÎã§Î©¥ Ï§ëÏïôÍ∞íÍ≥ºÎèÑ ÎπÑÍµêÌï¥ Î¥ÖÏãúÎã§.

```python
import numpy as np
import pandas as pd

np.random.seed(2025)

#1Î∞òÍ≥º 2Î∞ò Ï†êÏàò ÎßåÎì§Í∏∞
class1 = np.random.normal(75, 8, 30)
class2 = np.concatenate([np.random.normal(75, 15, 28),[0, 130]])

# Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùò ÌòïÌÉúÎ°ú ÌïôÏÉùÏùò Ï†êÏàò Ï†ÄÏû•ÌïòÍ∏∞
df3 = pd.DataFrame({'class': ['1Î∞ò'] * 30 + ['2Î∞ò'] * 30,'score': np.concatenate([class1, class2])})

# 2Î∞ò Ï†êÏàòÎßå Í∫ºÎÇ¥Ïò§Í∏∞
class2_scores = df3[df3['class'] == '2Î∞ò']['score']

# Q1, Q3, IQR Íµ¨ÌïòÍ∏∞
Q1 = class2_scores.quantile(0.25) # Q1: ÏïÑÎûòÏóêÏÑú 25% ÏúÑÏπò Ï†êÏàò
Q3 = class2_scores.quantile(0.75) # Q3: ÏúÑÏóêÏÑú 25% ÏúÑÏπò Ï†êÏàò
IQR = Q3 - Q1 # IQR = Q3 - Q1: Í∞ÄÏö¥Îç∞ 50% ÌïôÏÉùÏùò Ï†êÏàò Î≤îÏúÑ

# Ïù¥ÏÉÅÍ∞í Í∏∞Ï§Ä
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Ïù¥ÏÉÅÌïòÏßÄ ÏïäÏùÄ Ï†êÏàòÎßå ÎÇ®Í∏∞Í∏∞
class2_no_outliers = class2_scores[
    (class2_scores >= lower_bound) & (class2_scores <= upper_bound)
]

# ÌèâÍ∑†Í≥º Ï§ëÏïôÍ∞í ÎπÑÍµêÌïòÍ∏∞
mean_before = class2_scores.mean()
mean_after = class2_no_outliers.mean()
median_before = class2_scores.median()
median_after = class2_no_outliers.median()

# Í≤∞Í≥º Ï∂úÎ†•ÌïòÍ∏∞
print("=== IQR Outlier Detection ===")
print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
print(f"Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")

print("\n=== Outlier Removal Effect ===")
print(f"Mean before removing outliers: {mean_before:.2f}")
print(f"Mean after removing outliers:  {mean_after:.2f}")

print(f"\nMedian before removing outliers: {median_before:.2f}")
print(f"Median after removing outliers:  {median_after:.2f}")

print("\nNumber of scores before:", len(class2_scores))
print("Number of scores after :", len(class2_no_outliers))
print("Outliers removed      :", len(class2_scores) - len(class2_no_outliers))

```
