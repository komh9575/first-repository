```python
#   ()
df = df.drop(columns=['cc_num'])
```

```python
#     
#          
#      
#  amt, log_amt, is_high_amount     \
#    ,   ,   
#        .

df['log_amt'] = np.log1p(df['amt'])
df['is_high_amount'] = (df['amt'] > 5000).astype(int)
```

```python
#         
#   ( )

df['distance'] = np.sqrt(
    (df['lat'] - df['merch_lat'])**2 +
    (df['long'] - df['merch_long'])**2)
```

```python
from sklearn.ensemble import RandomForestClassifier

#is_fraud      
model = RandomForestClassifier(class_weight='balanced')
```

```python
# Columns to be dropped directly, not dummified (e.g., identifiers, dates)
cols_to_drop = ['dob', 'trans_num']

# Drop columns if they exist in df
for col in cols_to_drop:
    if col in df.columns:
        df = df.drop(columns=[col])

# Get a list of all current object columns in df that should be dummified
# Exclude 'trans_date_trans_time' as it will be dropped when creating X, not dummified here
object_cols_for_dummies = [col for col in df.select_dtypes(include='object').columns if col != 'trans_date_trans_time']

# Apply get_dummies to these columns if any exist
if object_cols_for_dummies:
    df = pd.get_dummies(df, columns=object_cols_for_dummies, drop_first=True)
```

```python
# x Y   x  is_fraud    feature 
# 'trans_date_trans_time'      
X = df.drop(['is_fraud', 'trans_date_trans_time'], axis=1)
y = df['is_fraud']
```

```python
#   ,    , ,    42
# y(label)  train test    
#  vs      
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_hat=knn.predict(X_test)
print(y_hat[0:10])
print(y_test.value[0:10])
```
