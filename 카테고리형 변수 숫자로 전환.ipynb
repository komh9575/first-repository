```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
cars["cars_title" ] = le.fit_transform(cars["title"])

le=LabelEncoder()
cars["cars_Fuel type" ] = le.fit_transform(cars["Fuel type"])

le=LabelEncoder()
cars["cars_Body type" ] = le.fit_transform(cars["Body type"])

le=LabelEncoder()
cars["cars_Engine" ] = le.fit_transform(cars["Engine"])

le=LabelEncoder()
cars["cars_Gearbox" ] = le.fit_transform(cars["Gearbox"])

le=LabelEncoder()
cars["cars_Emission Class" ] = le.fit_transform(cars["Emission Class"])

le=LabelEncoder()
cars["cars_Service history" ] = le.fit_transform(cars["Service history"])

le=LabelEncoder()
cars["cars_brand_name" ] = le.fit_transform(cars["brand_name"])
```

```python
cars = pd.get_dummies(cars, columns=["cars_title", "cars_Fuel type", "cars_Body type", "cars_Engine", "cars_Gearbox", "cars_Emission Class", "cars_Service history", "cars_brand_name"], drop_first=True)
```
