```python
# gender: None 값을 '알 수 없음'으로 변경
profile_df['gender'].fillna('알 수 없음', inplace=True)

# 소문자/대문자, 공백 등 이상값 제거 및 표준화 (예: 모두 대문자로 변환)
profile_df['gender'] = profile_df['gender'].astype(str).str.strip().str.upper()

# 범주형 변수로 명확히 고정
profile_df['gender'] = pd.Categorical(profile_df['gender'])

# 변경된 gender 값 확인
print("Gender 값 분포:")
display(profile_df['gender'].value_counts())
print("Gender 데이터 타입:", profile_df['gender'].dtype)
```

```python
import numpy as np

# age 118을 NaN으로 변경
profile_df['age'] = profile_df['age'].replace(118, np.nan)

# 연령대 구간 정의
bins = [18, 30, 40, 50, 60, profile_df['age'].max() + 1]
labels = ['18-29', '30-39', '40-49', '50-59', '60+']

# age_group 컬럼 생성
profile_df['age_group'] = pd.cut(profile_df['age'], bins=bins, labels=labels, right=False)

# 변경된 age 및 age_group 값 확인
print("Age 값에 118이 NaN으로 변경되었는지 확인 (처음 5개 행):")
display(profile_df[['age', 'age_group']].head())

print("\nAge_group 분포:")
display(profile_df['age_group'].value_counts().sort_index())
```

```python
print(f"Original profile_df shape: {profile_df.shape}")

# age가 NaN인 행 삭제
profile_df.dropna(subset=['age'], inplace=True)

print(f"New profile_df shape after dropping NaN ages: {profile_df.shape}")
display(profile_df.head())
```

```python
# became_member_on -> datetime으로 변경
profile_df['became_member_on'] = pd.to_datetime(profile_df['became_member_on'], format='%Y%m%d')

# 기준일 설정 (데이터 내 가장 최근 가입일 또는 분석 기준일)
# 여기서는 데이터 내 가장 최근 가입일을 기준으로 설정
reference_date = profile_df['became_member_on'].max()

# membership_days 파생 변수 추가
profile_df['membership_days'] = (reference_date - profile_df['became_member_on']).dt.days

# membership_days를 구간화한 변수 생성
bins_membership = [-1, 365, 365*3, profile_df['membership_days'].max() + 1]
labels_membership = ['<1년', '1-3년', '3년 이상']
profile_df['membership_group'] = pd.cut(profile_df['membership_days'], bins=bins_membership, labels=labels_membership, right=False)

# membership_days 이상치 처리 (상위 1% 또는 99% 기준으로 클리핑)
upper_bound = profile_df['membership_days'].quantile(0.99)
profile_df['membership_days'] = np.where(profile_df['membership_days'] > upper_bound, upper_bound, profile_df['membership_days'])

# 변경된 컬럼 확인
print("Became_member_on, membership_days, membership_group 확인 (처음 5개 행):")
display(profile_df[['became_member_on', 'membership_days', 'membership_group']].head())

print("\nMembership Group 분포:")
display(profile_df['membership_group'].value_counts().sort_index())
```

## Transaction 데이터만 추출 (구매 기준)

```python
# transaction 이벤트만 필터링
transaction_df = profile_transcript_df[
    profile_transcript_df['event'] == 'transaction'
].copy()

print(transaction_df.shape)
transaction_df.head()
```

## RFM 지표 계산

```python
snapshot_time = transaction_df['time'].max()

rfm_df = transaction_df.groupby('person_id').agg({
    'time': lambda x: snapshot_time - x.max(),  # Recency
    'person_id': 'count',                       # Frequency
    'amount': 'sum'                             # Monetary
}).rename(columns={
    'time': 'recency',
    'person_id': 'frequency',
    'amount': 'monetary'
}).reset_index()

rfm_df.head()
```

## 쿠폰(Offer) 행동 Feature 생성
- 쿠폰을 받는가?
- 쿠폰을 열어보는가?
- 쿠폰을 사용하는가?

```python

# 전체 로그 profile_transcript_df 중에서 쿠폰과 관련된 행동 필터링
# offer_received 쿠폰을 받음
# offer viewed 쿠폰을 열어봄
# offer completed 쿠폰을 사용함
# 구매가 아니라 쿠폰 반응 로그만 따로 모은 데이터셋

offer_df = profile_transcript_df[
    profile_transcript_df['event'].isin(['offer received', 'offer viewed', 'offer completed'])
].copy()


# 피벗 테이블로 고객 단위 집계
# 고객 한 명당 쿠폰 행동 횟수를 요약
# 머신러닝에서 바로 넣을 수 있는 숫자형 feature를 만드는 과정

offer_features = offer_df.pivot_table(
    index='person_id',
    columns='event',
    values='offer_id',
    aggfunc='count',
    fill_value=0
).reset_index()
offer_features.head()
```

```python
offer_df.head()
```

```python
# 받은 쿠폰 중 몇 %를 열어봤는가?

offer_features['offer_view_rate'] = (
    offer_features.get('offer viewed', 0) /
    offer_features.get('offer received', 1)
)

# 열어본 쿠폰 중 실제 사용한 비율 계산
offer_features['offer_complete_rate'] = (
    offer_features.get('offer completed', 0) /
    offer_features.get('offer viewed', 1)
)

offer_features.head()
```

```python
filtered_offers = offer_features[
    (offer_features['offer completed'] != 0) |
    (offer_features['offer_complete_rate'] != 0)
]
display(filtered_offers.head())
```

```python
zero_completed_offers_count = offer_features[
    (offer_features['offer completed'] == 0) &
    (offer_features['offer_complete_rate'] == 0)
].shape[0]

print(f"'offer completed'와 'offer_complete_rate'가 모두 0인 행의 개수: {zero_completed_offers_count}")
```

## 스케일링
- (값 - 평균) / 표준편차로 변경해줌
- 자기 변수 안에서 평균보다 얼마나 다른지 확인
- 구매가치 + 행동성향이 반영된 분석을 하고 싶기에, 스케일링 진행


```python

from sklearn.preprocessing import StandardScaler

feature_cols = [
    'recency', 'frequency', 'monetary',
    'offer received', 'offer viewed', 'offer completed',
    'offer_view_rate', 'offer_complete_rate'
]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_features[feature_cols])
```

## K-Means 클러스터링
- 고객유형을 4가지로 구분
- 랜덤초기값 10번 시도

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

customer_features['cluster'] = kmeans.fit_predict(scaled_features)

customer_features['cluster'].value_counts()
```

```python
cluster_summary = customer_features.groupby('cluster')[feature_cols].mean()
cluster_summary
```

## 코호트 분석
- 같은 시점에 들어온 고객들은 시간이 지나도 비슷한 패턴을 보이고 있는가?


```python
# 고객별 첫 구매 시점
first_purchase = transaction_df.groupby('person_id')['time'].min().reset_index()
first_purchase.columns = ['person_id', 'first_purchase_time']

transaction_cohort = transaction_df.merge(
    first_purchase,
    on='person_id',
    how='left'
)

# 코호트 기간 (주 단위 or 월 단위)
transaction_cohort['cohort_week'] = (
    transaction_cohort['first_purchase_time'] // 7
)

transaction_cohort['event_week'] = (
    transaction_cohort['time'] // 7
)

transaction_cohort['cohort_index'] = (
    transaction_cohort['event_week'] -
    transaction_cohort['cohort_week']
)
```
