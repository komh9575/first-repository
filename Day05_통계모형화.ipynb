## ë¬¸ì œ 1 (ë‚œì´ë„: í•˜): ë‹¤ì¤‘íšŒê·€ë¶„ì„ì˜ ì²«ê±¸ìŒ

> **ğŸ¯ ëª©í‘œ:** ì—¬ëŸ¬ ê°œì˜ ì›ì¸(ë…ë¦½ë³€ìˆ˜)ì„ ë™ì‹œì— ê³ ë ¤í•˜ì—¬ ê²°ê³¼ë¥¼ ì˜ˆì¸¡í•˜ëŠ” **ë‹¤ì¤‘íšŒê·€ë¶„ì„**ì„ íŒŒì´ì¬ ë°°ì—´ ë°©ì‹ìœ¼ë¡œ êµ¬í˜„í•©ë‹ˆë‹¤.


### ğŸ’¡ í•µì‹¬ ê°œë…:

ë‹¨ìˆœíšŒê·€ê°€ í•˜ë‚˜ì˜ ì›ì¸(`X`)ìœ¼ë¡œ ê²°ê³¼ë¥¼ ì„¤ëª…í–ˆë‹¤ë©´, **ë‹¤ì¤‘íšŒê·€**ëŠ” **ì—¬ëŸ¬ ê°œì˜ ì›ì¸(`X1, X2, ...`)**ì„ ë™ì‹œì— ê³ ë ¤í•˜ì—¬ ê²°ê³¼ë¥¼ ë” ì •êµí•˜ê²Œ ì„¤ëª…í•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤. ì´ë•Œ ê° ì›ì¸ì˜ ì˜í–¥ë ¥(ê³„ìˆ˜)ì€ **'ë‹¤ë¥¸ ì›ì¸ë“¤ì´ ëª¨ë‘ ë™ì¼í•˜ë‹¤ê³  í†µì œí–ˆì„ ë•Œ'**ì˜ ìˆœìˆ˜í•œ ì˜í–¥ë ¥ì„ ì˜ë¯¸í•©ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1. `tips` ë°ì´í„°ì…‹ì—ì„œ `tip`ì„ ì¢…ì†ë³€ìˆ˜(y)ë¡œ, `total_bill`ê³¼ `size`ë¥¼ ë…ë¦½ë³€ìˆ˜(X)ë¡œ ë¶„ë¦¬í•˜ì„¸ìš”.
2. ë…ë¦½ë³€ìˆ˜(X)ì— ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ì„¸ìš”.
3. `sm.OLS`ë¥¼ ì‚¬ìš©í•˜ì—¬ ë‹¤ì¤‘íšŒê·€ ëª¨ë¸ì„ í•™ìŠµì‹œí‚¤ê³ , `.summary()`ë¡œ ê²°ê³¼ë¥¼ í™•ì¸í•˜ì„¸ìš”.
4. ê²°ê³¼í‘œë¥¼ ë³´ê³  `total_bill`ê³¼ `size`ì˜ ê³„ìˆ˜(coef)ê°€ ê°ê° ë¬´ì—‡ì„ ì˜ë¯¸í•˜ëŠ”ì§€ í•´ì„í•´ ë³´ì„¸ìš”.


### âœï¸ ì½”ë“œ ì‘ì„±:


```python
import statsmodels.api as sm

# 1. ì—¬ê¸°ì— ë°ì´í„°ë¥¼ yì™€ Xë¡œ ë¶„ë¦¬í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# ì¢…ì†ë³€ìˆ˜ yì— 'tip' ì»¬ëŸ¼ì„ í• ë‹¹í•˜ì„¸ìš”.
y = tips['tip']
# íŒì„ ì•Œê³  ì‹¶ì€ ê²ƒ.
# ê³„ì‚°ê¸ˆì•¡ì´ ë§ìœ¼ë©´ íŒì„ ë” ì£¼ëŠ”ê°€?
# ì‚¬ì´ì¦ˆê°€ í¬ë©´ íŒì„ ë” ì£¼ëŠ”ê°€?

# ë…ë¦½ë³€ìˆ˜ Xì— 'total_bill'ê³¼ 'size' ì»¬ëŸ¼ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë¬¶ì–´ í• ë‹¹í•˜ì„¸ìš”.
X = tips[['total_bill', 'size']]

# 2. ì—¬ê¸°ì— ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# sm.add_constant()ë¥¼ ì‚¬ìš©í•´ X ë°ì´í„°ì— ìƒìˆ˜í•­ì„ ì¶”ê°€í•©ë‹ˆë‹¤.
X_const = sm.add_constant(X)

# 3. ì—¬ê¸°ì— ëª¨ë¸ì„ í•™ìŠµí•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# sm.OLS() í•¨ìˆ˜ì— yì™€ X_constë¥¼ ìˆœì„œëŒ€ë¡œ ë„£ì–´ ëª¨ë¸ì„ ë§Œë“¤ê³  í•™ìŠµì‹œí‚µë‹ˆë‹¤.
model_1 = sm.OLS(y, X_const).fit()
print(model_1.summary())

# 4. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# íšŒê·€ì‹: tip=0.6689+0.0927â‹…total_bill+0.1926â‹…size
# ê¸°ìš¸ê¸° =â€œXê°€ 1ë§Œí¼ ë³€í•  ë•Œ, YëŠ” ì–¼ë§ˆë‚˜ ë³€í•˜ë‚˜â€
# total_billì™€ sizeì˜ ê°’ì„ 0ìœ¼ë¡œ ì„¤ì •í–ˆì„ ë•Œ, ì˜ˆìƒë˜ëŠ” í‰ê· ì˜ íŒì€ ì•½ 0.67ë‹¬ëŸ¬ê°€ ëœë‹¤. (const : 0.6689)/ ëª¨ë“  ë…ë¦½ë³€ìˆ˜ê°€ 0ì¼ ë•Œì˜ ì¢…ì†ë³€ìˆ˜ë¥¼ ì˜ˆìƒí•˜ëŠ” ê°’

# total_bill ê³„ìˆ˜ì˜ ì˜ë¯¸: 0.0927ì´ ë‚˜ì™”ìŒ. í…Œì´ë¸” ì¸ì›ìˆ˜(size)ê°€ ë™ì¼í•˜ë‹¤ê³  ê°€ì •í–ˆì„ ë•Œ, ì´ ê³„ì‚°ê¸ˆì•¡ì´ 1ë‹¬ëŸ¬ ì¦ê°€í•˜ë©´ íŒì€ í‰ê· ì ìœ¼ë¡œ ì•½ 0.093ë‹¬ëŸ¬ ì¦ê°€í•œë‹¤.
# size ê³„ìˆ˜ì˜ ì˜ë¯¸: 0.1926 totall billì´ ë™ì¼í•˜ë‹¤ê³  ê°€ì •í–ˆì„ ë•Œ, í…Œì´ë¸” ì¸ì›ìˆ˜(size)ê°€ 1ëª… ì¦ê°€í•˜ë©´ íŒ(tip)ì€ 0.1926 ì¦ê°€í•œë‹¤.
```

### ğŸ¤” ìƒê°í•´ ë³¼ ë¬¸ì œ:

### ğŸ’¡ í•µì‹¬ ê°œë…:

**ë”ë¯¸ ë³€ìˆ˜**ëŠ” 'ì„±ë³„', 'ìš”ì¼'ê³¼ ê°™ì€ ë¬¸ì(ë²”ì£¼í˜•) ë°ì´í„°ë¥¼ í†µê³„ ëª¨ë¸ì— ì‚¬ìš©í•˜ê¸° ìœ„í•´ **0ê³¼ 1ë¡œ ì´ë£¨ì–´ì§„ ìˆ«ì ë³€ìˆ˜**ë¡œ ë³€í™˜í•˜ëŠ” ê¸°ë²•ì…ë‹ˆë‹¤. Pandasì˜ `pd.get_dummies()` í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ë©´ ì´ ê³¼ì •ì„ ì‰½ê²Œ ì²˜ë¦¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ë•Œ, ì—¬ëŸ¬ ë²”ì£¼ ì¤‘ í•˜ë‚˜ëŠ” **ê¸°ì¤€(reference)**ì´ ë˜ì–´ ë¶„ì„ ì˜¤ë¥˜ë¥¼ ë§‰ê³  í•´ì„ì˜ ê¸°ì¤€ì  ì—­í• ì„ í•©ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1. `total_bill`ê³¼ 'ìš”ì¼(`day`)'ì´ íŒ(`tip`)ì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì„ ë¶„ì„í•´ë´…ì‹œë‹¤.
2. `pd.get_dummies()`ë¥¼ ì‚¬ìš©í•˜ì—¬ 'day' ì»¬ëŸ¼ì„ ë”ë¯¸ ë³€ìˆ˜ë¡œ ë³€í™˜í•˜ì„¸ìš”. (ë‹¨, `drop_first=True` ì˜µì…˜ ì‚¬ìš©)
3. ê¸°ì¡´ ë…ë¦½ë³€ìˆ˜(`total_bill`)ì™€ ìƒì„±ëœ ë”ë¯¸ ë³€ìˆ˜ë¥¼ í•©ì³ ìµœì¢… X ë°ì´í„°ë¥¼ ë§Œë“œì„¸ìš”.
4. ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ê³  íšŒê·€ ëª¨ë¸ì„ í•™ìŠµì‹œí‚¨ í›„, ê²°ê³¼ë¥¼ í•´ì„í•˜ì„¸ìš”.


### âœï¸ ì½”ë“œ ì‘ì„±:


```python

# 1. ê¸°ë³¸ ë°ì´í„°ë¥¼ ë¶„ë¦¬í•©ë‹ˆë‹¤.
y = tips['tip']
X_base = tips[['total_bill']]

# 2. ì—¬ê¸°ì— 'day' ì»¬ëŸ¼ìœ¼ë¡œ ë”ë¯¸ ë³€ìˆ˜ë¥¼ ìƒì„±í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# pd.get_dummies()ë¥¼ ì‚¬ìš©í•˜ê³ , drop_first=True ì˜µì…˜ì„ ìŠì§€ ë§ˆì„¸ìš”.
day_dummies = pd.get_dummies(tips['day'], prefix='day', drop_first=True, dtype=int)

# 3. ì—¬ê¸°ì— ê¸°ë³¸ X ë°ì´í„°ì™€ ë”ë¯¸ ë³€ìˆ˜ë¥¼ í•©ì¹˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# pd.concatì„ ì‚¬ìš©í•˜ê³ , axis=1 ì˜µì…˜ìœ¼ë¡œ ì˜†ìœ¼ë¡œ ë¶™ì…ë‹ˆë‹¤.
X = pd.concat([X_base, day_dummies], axis=1)

# 4. ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ê³  ëª¨ë¸ì„ í•™ìŠµì‹œí‚¨ í›„ ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ì„¸ìš”.
X_const = sm.add_constant(X)
model_2 = sm.OLS(y, X_const).fit()
print(model_2.summary())

# tip=Î²0â€‹+Î²1â€‹â‹…total_bill+Î²2â€‹â‹…day_Sat+â‹¯

# 5. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# ê¸°ì¤€ì´ ëœ ìš”ì¼ì€ ë¬´ì—‡ì¸ê°€?: ëª©ìš”ì¼ì´ ê¸°ì¤€ì´ ë˜ì–´ ê¸ˆ, í† , ì¼ì„ ì¸¡ì •í•œë‹¤.
# 'day_Sat' ê³„ìˆ˜ì˜ ì˜ë¯¸: day_Satì˜ ê³„ìˆ˜ëŠ” -0.0671ì´ë‹¤. ì´ ê³„ì‚°ê¸ˆì•¡ì´ ë™ì¼í•˜ë‹¤ê³  ë³´ì•˜ì„ ë•Œ, í† ìš”ì¼ì— ë°›ì€ íŒì€ ëª©ìš”ì¼ì— ë¹„í•´ í‰ê· ì ìœ¼ë¡œ ì•½ 0.067ë‹¬ëŸ¬ ì ë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•œë‹¤.
# ìš”ì¼ ë”ë¯¸ ë³€ìˆ˜ì˜ ê³„ìˆ˜ëŠ” ì†Œë¹„ ê·œëª¨(total_bill)ì˜ ì˜í–¥ì„ ì œê±°í•œ ìƒíƒœì—ì„œ, ìš”ì¼ ê·¸ ìì²´ê°€ íŒì— ë¯¸ì¹˜ëŠ” ìˆœìˆ˜í•œ ì°¨ì´ë¥¼ ì˜ë¯¸í•˜ê¸° ë•Œë¬¸ì— ì´ ê³„ì‚°ê¸ˆì•¡ì„ ë™ì¼í•˜ê²Œ ë‘ê³  í•´ì„í•œë‹¤.
```

aê°€ yì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì´ b(ìƒí™©)ì— ë”°ë¼ ì„¸ì§€ê¸°ë„, ì•½í•´ì§€ê¸°ë„ í•˜ë©´ ìƒí˜¸ì‘ìš©ì´ ìˆë‹¤ê³  íŒë‹¨í•  ìˆ˜ ìˆë‹¤.

### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1. "ì‹ì‚¬ ì´ì•¡(`total_bill`)ì´ íŒì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì€ í¡ì—°ì(`smoker`) ê·¸ë£¹ê³¼ ë¹„í¡ì—°ì ê·¸ë£¹ì—ì„œ ë‹¤ë¥¼ ê²ƒì´ë‹¤"ë¼ëŠ” ê°€ì„¤ì„ ê²€ì¦í•´ë´…ì‹œë‹¤.
2. `smoker` ë³€ìˆ˜ë¥¼ ë”ë¯¸ ë³€ìˆ˜(`smoker_Yes`)ë¡œ ë§Œë“œì„¸ìš”. ('Yes' = 1, 'No' = 0)
3. `total_bill`ê³¼ `smoker_Yes`ë¥¼ ê³±í•˜ì—¬ ìƒí˜¸ì‘ìš© í•­(`bill_x_smoker`)ì„ ë§Œë“œì„¸ìš”.
4. `total_bill`, `smoker_Yes`, `bill_x_smoker`ë¥¼ ëª¨ë‘ í¬í•¨í•˜ì—¬ íšŒê·€ ëª¨ë¸ì„ í•™ìŠµì‹œí‚¤ê³  ê²°ê³¼ë¥¼ í•´ì„í•˜ì„¸ìš”.


### âœï¸ ì½”ë“œ ì‘ì„±:


```python
# 1. ê¸°ë³¸ ë°ì´í„°ë¥¼ ì¤€ë¹„í•©ë‹ˆë‹¤.
y = tips['tip']
X = tips[['total_bill']]

# 2. ì—¬ê¸°ì— 'smoker' ë”ë¯¸ ë³€ìˆ˜ë¥¼ Xì— ì¶”ê°€í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
X['smoker_Yes'] = pd.get_dummies(tips['smoker'], drop_first=True, dtype=int)

# 3. ì—¬ê¸°ì— ìƒí˜¸ì‘ìš© í•­ì„ ì§ì ‘ ê³„ì‚°í•˜ì—¬ Xì— ì¶”ê°€í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# 'total_bill' ì»¬ëŸ¼ê³¼ 'smoker_Yes' ì»¬ëŸ¼ì„ ê³±í•©ë‹ˆë‹¤.
X['bill_x_smoker'] = X['total_bill'] * X['smoker_Yes']

# 4. ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ê³  ëª¨ë¸ì„ í•™ìŠµì‹œí‚¨ í›„ ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ì„¸ìš”.
X_const = sm.add_constant(X)
model_3 = sm.OLS(y, X_const).fit()
print(model_3.summary())

# 5. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# ìƒí˜¸ì‘ìš© í•­(bill_x_smoker)ì˜ p-valueëŠ” ìœ ì˜ë¯¸í•œê°€?: bill_x_smokerì˜ p-valueëŠ” 0.000ìœ¼ë¡œ 0.05ë³´ë‹¤ ì‘ë‹¤. ì´ì— í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•˜ë‹¤ê³  í•  ìˆ˜ ìˆë‹¤.
# ì´ëŠ” í¡ì—° ì—¬ë¶€ì— ë”°ë¼ total billì— ë¯¸ì¹˜ëŠ” íš¨ê³¼ê°€ ë‹¤ë¥´ë‹¤ëŠ” ê²ƒì´ë‹¤. (ìƒí˜¸ì‘ìš©ì´ ì‹¤ì œë¡œ ì¡´ì¬í•œë‹¤ëŠ” ëœ»)
# ìƒí˜¸ì‘ìš© í•­ ê³„ìˆ˜ì˜ ì˜ë¯¸:
# bill_x_smoker 0.0676 ì˜ ìˆ˜ì¹˜ë¥¼ ë³´ì´ê³  ìˆìœ¼ë¯€ë¡œ, í¡ì—°ìì¼ ë•Œ total billì´ íŒì— ë¯¸ì¹˜ëŠ” íš¨ê³¼ëŠ” ë¹„í¡ì—°ìì— ë¹„í•´ 1ë‹¬ëŸ¬ë‹¹ ì•½ 0.0676ë‹¬ëŸ¬ë§Œí¼ ë” í¬ë‹¤ëŠ” ê²ƒì„ ì•Œ ìˆ˜ ìˆë‹¤.
```

### ğŸ¤” ìƒê°í•´ ë³¼ ë¬¸ì œ:

## ë¬¸ì œ 4 (ë‚œì´ë„: ì¤‘): ë¡œì§€ìŠ¤í‹± íšŒê·€ë¶„ì„

> **ğŸ¯ ëª©í‘œ:** ê²°ê³¼ê°€ 'ì˜ˆ/ì•„ë‹ˆì˜¤' ê°™ì€ ë²”ì£¼í˜•ì¼ ë•Œ ì‚¬ìš©í•˜ëŠ” **ë¡œì§€ìŠ¤í‹± íšŒê·€ë¶„ì„**ì„ êµ¬í˜„í•˜ê³ , ê·¸ ê²°ê³¼ë¥¼ **ì˜¤ì¦ˆë¹„(Odds Ratio)**ë¡œ í•´ì„í•˜ëŠ” ë°©ë²•ì„ ë°°ì›ë‹ˆë‹¤.


### ğŸ’¡ í•µì‹¬ ê°œë…:

**ë¡œì§€ìŠ¤í‹± íšŒê·€ë¶„ì„**ì€ ê²°ê³¼ê°€ 'ì˜ˆ/ì•„ë‹ˆì˜¤', 'ì„±ê³µ/ì‹¤íŒ¨'ì²˜ëŸ¼ **ë‘ ê°€ì§€ ì¤‘ í•˜ë‚˜**ì¼ ë•Œ ì‚¬ìš©í•˜ëŠ” ë¶„ì„ ë°©ë²•ì…ë‹ˆë‹¤. ì–´ë–¤ ì¡°ê±´ì—ì„œ íŠ¹ì • ê²°ê³¼ê°€ ë‚˜íƒ€ë‚  **'í™•ë¥ '**ì„ ì˜ˆì¸¡í•©ë‹ˆë‹¤. íšŒê·€ ê³„ìˆ˜ëŠ” ê·¸ëŒ€ë¡œ í•´ì„í•˜ê¸° ì–´ë µê³ , `np.exp(ê³„ìˆ˜)`ë¥¼ ê³„ì‚°í•œ **ì˜¤ì¦ˆë¹„(Odds Ratio)**ë¡œ í•´ì„í•´ì•¼ í•©ë‹ˆë‹¤. ì˜¤ì¦ˆë¹„ê°€ 1ë³´ë‹¤ í¬ë©´ í™•ë¥  ì¦ê°€, 1ë³´ë‹¤ ì‘ìœ¼ë©´ í™•ë¥  ê°ì†Œë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1. ì‹ì‚¬ ì´ì•¡(`total_bill`)ê³¼ íŒ(`tip`) ê¸ˆì•¡ì„ ë³´ê³ , ê·¸ ê³ ê°ì´ 'ë‚¨ì„±'ì¼ í™•ë¥ ì„ ì˜ˆì¸¡í•˜ëŠ” ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ì„ ë§Œë“¤ì–´ë´…ì‹œë‹¤.
2. ì¢…ì†ë³€ìˆ˜ `y`ë¥¼ 'sex'ê°€ 'Male'ì´ë©´ 1, ì•„ë‹ˆë©´ 0ì´ ë˜ë„ë¡ ë³€í™˜í•˜ì„¸ìš”.
3. `sm.Logit` í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ì„ í•™ìŠµì‹œí‚¤ê³ , ê²°ê³¼ë¥¼ í™•ì¸í•˜ì„¸ìš”.
4. `np.exp()`ë¥¼ ì´ìš©í•´ ê³„ìˆ˜ë¥¼ ì˜¤ì¦ˆë¹„ë¡œ ë³€í™˜í•˜ê³ , ê·¸ ì˜ë¯¸ë¥¼ í•´ì„í•˜ì„¸ìš”.


### âœï¸ ì½”ë“œ ì‘ì„±:


```python
# 1. ì—¬ê¸°ì— ì¢…ì†ë³€ìˆ˜ yë¥¼ 0ê³¼ 1ë¡œ ë³€í™˜í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# 'sex'ê°€ 'Male'ì´ë©´ 1, ì•„ë‹ˆë©´ 0
y = tips['sex'].apply(lambda x: 1 if x == 'Male' else 0)

# 2. ë…ë¦½ë³€ìˆ˜ Xë¥¼ ì¤€ë¹„í•˜ê³  ìƒìˆ˜í•­ì„ ì¶”ê°€í•˜ì„¸ìš”.
X = tips[['total_bill', 'tip']]
X_const = sm.add_constant(X)

# 3. ì—¬ê¸°ì— ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ì„ í•™ìŠµì‹œí‚¤ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# sm.Logit() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
model_4 = sm.Logit(y, X_const).fit()
print(model_4.summary())

# 4. ì—¬ê¸°ì— ì˜¤ì¦ˆë¹„ë¥¼ ê³„ì‚°í•˜ê³  ì¶œë ¥í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
odds_ratios = np.exp(model_4.params)
print("\n--- ì˜¤ì¦ˆë¹„ (Odds Ratios) ---")
print(odds_ratios)

# 5. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# total_bill ì˜¤ì¦ˆë¹„ì˜ ì˜ë¯¸:
# total_billì˜ ì˜¤ì¦ˆë¹„ = 1.0408
# ë¡œì§€ìŠ¤í‹± íšŒê·€ëŠ” 1=ë‚¨ì„±ìœ¼ë¡œ ì½”ë”©í•œ ê²°ê³¼ì˜ í™•ë¥ (ì˜¤ì¦ˆ)ë¥¼ ë„ì¶œ
# "ì´ ê³„ì‚° ê¸ˆì•¡ì´ 1ë‹¬ëŸ¬ ì¦ê°€í•  ë•Œ ë‚¨ì„±ì¼ ì˜¤ì¦ˆ(í™•ë¥ )ì€ ì•½ 1.04ë°° ì¦ê°€í•œë‹¤."

# tip ì˜¤ì¦ˆë¹„ì˜ ì˜ë¯¸:
# "ì´ íŒì˜ ê¸ˆì•¡ì´ 1ë‹¬ëŸ¬ ì¦ê°€í•  ë•Œ ë‚¨ì„±ì¼ ì˜¤ì¦ˆ(í™•ë¥ )ì€ ì•½ 0.97ë°°ë¡œ ê°ì†Œí•œë‹¤."
# ì˜¤ì¦ˆë¹„ 1ë³´ë‹¤ ì‘ìœ¼ë¯€ë¡œ (0.972 < 1) ê°ì†Œí•œë‹¤ê³  í‘œí˜„
```

### ğŸ¤” ìƒê°í•´ ë³¼ ë¬¸ì œ:

`model_4.predict(X_const)` ì½”ë“œë¥¼ ì‹¤í–‰í•˜ë©´ ê° ì†ë‹˜ì— ëŒ€í•´ 'ë‚¨ì„±ì¼ í™•ë¥ 'ì„ ì˜ˆì¸¡í•´ì¤ë‹ˆë‹¤. ì´ ì˜ˆì¸¡ í™•ë¥ ì´ ê°€ì¥ ë†’ì€ ìƒìœ„ 5ëª…ì˜ ì†ë‹˜ ë°ì´í„°(total_bill, tip)ëŠ” ì–´ë–¤ íŠ¹ì§•ì„ ê°€ì§€ê³  ìˆëŠ”ì§€ í™•ì¸í•´ë³´ì„¸ìš”.


```python
# ê° ì†ë‹˜ì— ëŒ€í•´ 'ë‚¨ì„±ì¼ í™•ë¥ ' ì˜ˆì¸¡
pred_prob = model_4.predict(X_const)

# ì›ë³¸ ë°ì´í„°ì— ì˜ˆì¸¡ í™•ë¥  ì¶”ê°€
tips_with_prob = tips.copy()
tips_with_prob['prob_male'] = pred_prob

top5 = tips_with_prob.sort_values('prob_male', ascending=False).head(5)
top5['tip_ratio'] = top5['tip'] / top5['total_bill']

top5[['total_bill', 'tip', 'tip_ratio', 'prob_male']]
```

## ë¬¸ì œ 5 (ë‚œì´ë„: ì¤‘): ìµœê³ ì˜ ëª¨ë¸ ì°¾ê¸°: AIC

> **ğŸ¯ ëª©í‘œ:** ì—¬ëŸ¬ ëª¨ë¸ ì¤‘ ì–´ë–¤ ëª¨ë¸ì´ ê°€ì¥ ì¢‹ì€ì§€ ê°ê´€ì ì¸ ì ìˆ˜ë¡œ í‰ê°€í•˜ëŠ” **AIC**ì˜ ê°œë…ì„ ì´í•´í•˜ê³ , ì´ë¥¼ ì´ìš©í•´ ìµœì ì˜ ëª¨ë¸ì„ ì„ íƒí•©ë‹ˆë‹¤.


### ğŸ’¡ í•µì‹¬ ê°œë…:

**AIC (Akaike Information Criterion)**ëŠ” ì—¬ëŸ¬ ëª¨ë¸ ì¤‘ ì–´ë–¤ ëª¨ë¸ì´ ê°€ì¥ ì¢‹ì€ì§€ í‰ê°€í•˜ëŠ” 'ì ìˆ˜'ì…ë‹ˆë‹¤. AICëŠ” **ëª¨ë¸ì˜ ì„¤ëª…ë ¥(ì–¼ë§ˆë‚˜ ì˜ ë§ì¶”ëŠ”ê°€)**ì€ ë†’ì´ê³ , **ëª¨ë¸ì˜ ë³µì¡ì„±(ì–¼ë§ˆë‚˜ ë§ì€ ë³€ìˆ˜ë¥¼ ì¼ëŠ”ê°€)**ì—ëŠ” ë²Œì ì„ ì¤ë‹ˆë‹¤. ë”°ë¼ì„œ **AIC ì ìˆ˜ëŠ” ë‚®ì„ìˆ˜ë¡ ë” ì¢‹ì€(íš¨ìœ¨ì ì¸) ëª¨ë¸**ì´ë¼ê³  í‰ê°€í•©ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1. íŒ(`tip`) ê¸ˆì•¡ì„ ì˜ˆì¸¡í•˜ê¸° ìœ„í•œ ì„¸ ê°€ì§€ ë‹¤ë¥¸ ëª¨ë¸ì„ ë§Œë“­ë‹ˆë‹¤.
   - ëª¨ë¸ A: `total_bill`ë§Œ ì‚¬ìš©
   - ëª¨ë¸ B: `total_bill`, `size` ì‚¬ìš©
   - ëª¨ë¸ C: `total_bill`, `size`, `day`(ë”ë¯¸) ì‚¬ìš©
2. ê° ëª¨ë¸ì˜ `.aic` ì†ì„±ì„ í™•ì¸í•˜ì—¬ AIC ì ìˆ˜ë¥¼ ë¹„êµí•˜ì„¸ìš”.
3. ì–´ë–¤ ëª¨ë¸ì´ ê°€ì¥ ì¢‹ì€ ëª¨ë¸ì¸ì§€ ì„ íƒí•˜ê³ , ê·¸ ì´ìœ ë¥¼ ì„¤ëª…í•˜ì„¸ìš”.


### âœï¸ ì½”ë“œ ì‘ì„±:


```python
y = tips['tip']

# --- ëª¨ë¸ A: total_billë§Œ ì‚¬ìš© ---
X_A = tips[['total_bill']]
X_A_const = sm.add_constant(X_A)
model_A = sm.OLS(y, X_A_const).fit()
print("Model A AIC:", model_A.aic)

# --- ëª¨ë¸ B: total_billê³¼ size ì‚¬ìš© ---
# 1. ì—¬ê¸°ì— ëª¨ë¸ Bì˜ X ë°ì´í„°ë¥¼ ì¤€ë¹„í•˜ê³ , í•™ìŠµ í›„ AICë¥¼ ì¶œë ¥í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
X_B = tips[['total_bill', 'size']]
X_B_const = sm.add_constant(X_B)
model_B = sm.OLS(y, X_B_const).fit()
print("Model B AIC:", model_B.aic)

# --- ëª¨ë¸ C: total_bill, size, day ì‚¬ìš© ---
# 2. ì—¬ê¸°ì— ëª¨ë¸ Cì˜ X ë°ì´í„°ë¥¼ ì¤€ë¹„í•˜ê³ , í•™ìŠµ í›„ AICë¥¼ ì¶œë ¥í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
# (íŒíŠ¸: ë¬¸ì œ 2ì—ì„œ X ë°ì´í„°ë¥¼ ë§Œë“¤ì—ˆë˜ ê³¼ì •ì„ ì°¸ê³ í•˜ì„¸ìš”.)
day_dummies = pd.get_dummies(tips['day'], prefix='day', drop_first=True, dtype=int)
X_C = pd.concat([X_B, day_dummies], axis=1)
X_C_const = sm.add_constant(X_C)
model_C = sm.OLS(y, X_C_const).fit()
print("Model C AIC:", model_C.aic)

# 3. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# ê°€ì¥ ì¢‹ì€ ëª¨ë¸(AICê°€ ê°€ì¥ ë‚®ì€ ëª¨ë¸)ì€?:Model B AIC: 701.9702091086953
# ê·¸ë ‡ê²Œ ìƒê°í•˜ëŠ” ì´ìœ :aicëŠ” ëª¨ë¸ì˜ ì í•©ë„ë¥¼ ì¢…í•©ì ìœ¼ë¡œ ê³ ë ¤í•˜ì—¬ í‰ê°€í•˜ëŠ” ì§€í‘œì´ë‹¤. 3ê°œì˜ ëª¨ë¸ ì¤‘ bì˜ ëª¨ë¸ ì§€í‘œ ìˆ˜ì¹˜ê°€ ê°€ì¥ ë‚®ë‹¤.
# ìš”ì¼ë³€ìˆ˜ë¥¼ ì¶”ê°€í•œ ëª¨ë¸ cì˜ ê²½ìš° ë³µì¡ë„ëŠ” ì¦ê°€í•˜ì˜€ìœ¼ë‚˜, ê·¸ê²ƒì— ë¹„í•˜ì—¬ ê°œì„ ì´ í¬ì§€ ì•Šì•˜ë˜ ë“¯ ë³´ì¸ë‹¤.
# ê·¸ ì´ìœ ëŠ” ì•ì„œ ë¶„ì„í•œ ëª¨ë¸ bë³´ë‹¤ aicì˜ ì§€í‘œê°€ í¬ê¸° ë•Œë¬¸ì´ë‹¤.
```

### ğŸ¤” ìƒê°í•´ ë³¼ ë¬¸ì œ:

AICëŠ” ëª¨ë¸ì˜ ë³µì¡ë„(ì‚¬ìš©í•œ ë³€ìˆ˜ì˜ ê°œìˆ˜)ì— ë²Œì ì„ ì¤ë‹ˆë‹¤. ëª¨ë¸ BëŠ” ëª¨ë¸ Aë³´ë‹¤ ë³€ìˆ˜ê°€ í•˜ë‚˜ ë” ë§ìŒì—ë„ AICê°€ ë‚®ì•„ì¡ŒìŠµë‹ˆë‹¤. ì´ëŠ” ë¬´ì—‡ì„ ì˜ë¯¸í• ê¹Œìš”?

= ëª¨ë¸ bì˜ ë³€ìˆ˜ê°€ í•˜ë‚˜ ë” ë§ì•„ ë³µì¡ë„ê°€ ì¦í•˜ì˜€ìœ¼ë‚˜, ì¢…í•©ì ìœ¼ë¡œ total_billê³¼ sizeë¥¼ ê³ ë ¤í•˜ì˜€ì„ ë•Œ, ì„¤ëª…ë ¥ ê°œì„ ì´ ë” ì¦ê°€í•˜ì˜€ìŒì„ ì˜ë¯¸í•œë‹¤.
