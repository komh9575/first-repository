```python
numeric_cols = trip_csv.select_dtypes(include=['int64', 'float64']).columns  #숫자로 된 dtypedmf 보함한 컬럼을 추출하고 numeric_cols 이름으로 저장
print(numeric_cols)  # numeric_cols을 프린트
```

```python
import matplotlib.pyplot as plt  # matplotlib.pyplot as plt이라는 도구를 사용할 것임
for col in numeric_cols: #numeric_cols에서 col이라는 값을 불러올 것
    plt.scatter(trip_csv.index, trip_csv[col]) #trip_csv의 데이터에서 col의 값과 그 외에 trip_csv의 행열을 인덱스 처리하여 scatter의 모양으로 출력
    plt.title(f"Scatter plot of {col}") #scatter의 타이틀의 제목을 col값으로 지정
    plt.xlabel("Index") #x축을 인덱스의 이름으로 지정
    plt.ylabel(col) #y축을 col값으로 지정
    plt.show() #출력
```

```python
for col in numeric_cols: #boxplot은 결정치의 값을 한 눈에 보여주는 데이터 그래프인데, x축을 기준으로 그 안에는 y축에 해당하는 값을 보여주어 결정치의 분포도를 확인
    sns.boxplot(x=trip_csv[col]) #trip_csv[col]을 x축에 출력
    plt.xlabel(col)
    plt.show()
```

```python
for col in numeric_cols:
  sns.boxplot(y=trip_csv[col])
  plt.ylabel(col)
  plt.show()
```

```python
cols = ['tpep_pickup_datetime',	'tpep_dropoff_datetime',	'payment_method', 'trip_distance',	'tolls_amount']  # 일단 내가 필요한 열만 추출
df_selected = trip_csv[cols]
df_selected
```

```python
trip_csv['tpep_pickup_datetime']=pd.to_datetime(trip_csv['tpep_pickup_datetime'], format='%m/%d/%Y %I:%M:%S %p')
trip_csv['tpep_dropoff_datetime']=pd.to_datetime(trip_csv['tpep_dropoff_datetime'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
trip_csv['trip_duration'] = trip_csv['tpep_dropoff_datetime'] - trip_csv['tpep_pickup_datetime']
trip_csv['trip_duration_min'] = trip_csv['trip_duration'].dt.total_seconds() / 60
print(trip_csv['trip_duration_min'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(trip_csv['trip_duration_min'], trip_csv['trip_distance'], alpha=1) #고유한 거리 조사하고 여행 거리를 x축과 y축으로 대입
plt.xlabel('trip_duration')
plt.ylabel('Frequency')
plt.title('Distribution of Trip Duration')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(trip_csv['tolls_amount'], trip_csv['trip_distance'], alpha=1) #고유한 거리 조사하고 여행 거리를 x축과 y축으로 대입
plt.xlabel('tolls_amount')
plt.ylabel('Frequency')
plt.title('Amount of Distance')
plt.show()
```
